```python
from pandas_datareader import data, wb
import yfinance as yf
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
pip install --upgrade pandas_datareader
```

```python
# BAC = data.DataReader('BAC','yahoo',start,end)
BAC = yf.download('BAC', start='2020-01-01', end='2023-01-01')
GS = yf.download('GS', start='2020-01-01', end='2023-01-01')
JPM = yf.download('JPM', start='2020-01-01', end='2023-01-01')
C = yf.download('C', start='2020-01-01', end='2023-01-01')
MS = yf.download('MS', start='2020-01-01', end='2023-01-01')
WFC = yf.download('WFC', start='2020-01-01', end='2023-01-01')
```

```python
BAC.head()
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
# bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Back Ticker', 'Stock Info']
```

```python
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

```python
sns.pairplot(returns[1:])
```

```python
returns.idxmin()
```

```python
returns.std()
```

```python
returns.loc['2020-01-01':'2020-12-31'].std()
```

```python
sns.distplot(returns.loc['2020-01-01':'2020-12-31']['MS Return'],color='pink',
            bins=50)
```

```python
sns.distplot(returns.loc['2020-01-01':'2020-12-31']['C Return'],color='red',bins=50)
```
